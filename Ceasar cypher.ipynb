```python
print("Caesar Cypher")

alphabet = [
'a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n', 'o', 'p', 'q', 'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z'
]

alphabet_cap = [let.upper() for let in alphabet]

def caesar(message, shift, sense):

    user_message_list = []

    if sense == "decode":
        shift = shift * -1

    for key in message:
        if key in alphabet:
            shifted_key = alphabet[(alphabet.index(key) + shift) % 26]
        elif key.isupper():
            shifted_key = alphabet_cap[(alphabet_cap.index(key) + shift) % 26]
        else:
            shifted_key = key
        user_message_list.append(shifted_key)

    return f"Coded message {''.join(user_message_list)}"

print(caesar("Kl Khoor!", 3, "decode"))
```
