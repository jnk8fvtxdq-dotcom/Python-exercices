```python
#***V2***
import random
cards = [11, 2, 3, 4, 5, 6, 7, 8, 9, 10, 10, 10, 10]

def results(player_score, computer_score, replay):
    if player_score == computer_score:
        outcome = "Draw"
    elif player_score > 21 or computer_score == 21:
        outcome = "Lose"
    elif computer_score > 21 or player_score == 21:
        outcome = "Win"
    elif replay == "n":
        if player_score > computer_score:
            outcome = "Win"
        else:
            outcome = "Lose"
    else:
        outcome = "Follow"
    return outcome

def format_results(player_score, computer_score, player_list, computer_list, outcome):
    if outcome == "Draw":
        return f"Your score: {player_score}\nComputer score: {computer_score}\nIt's a draw!"
    elif outcome == "Lose":
        return f"Your score: {player_score}\nComputer score: {computer_score}\nYou lose!"
    elif outcome == "Win":
        return f"Your score: {player_score}\nComputer score: {computer_score}\nYou win!"
    elif outcome == "Follow":
        return f"Your cards: {player_list}, current score: {player_score}\nComputer's first card: {computer_list[0]}"

def draw_cards(cards, player_list, computer_list, play_count):
    player_list.append(random.choice(cards))
    if play_count == 0:
        player_list.append(random.choice(cards))
    computer_list.append(random.choice(cards))

while True:
    play = input("Do you want to play a game of Blackjack? Type 'y' or 'n':")
    if play == "y":
        player_list = []
        computer_list = []
        play_count = 0
        replay = "y"
        while replay == "y":
            draw_cards(cards, player_list, computer_list, play_count)
            play_count += 1
            player_score = sum(player_list)
            computer_score = sum(computer_list)
            outcome = results(player_score, computer_score, replay)
            print(format_results(player_score, computer_score, player_list, computer_list, outcome))
            if outcome != "Follow":
                break
            replay = input("Type 'y' to get another card, type 'n' to pass:")
            if replay == 'n':
                outcome = results(player_score, computer_score, replay)
                print(format_results(player_score, computer_score, player_list, computer_list, outcome))
            if outcome != "Follow":
                break
    if play == "n":
        break
```
