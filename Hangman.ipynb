Hangman

```python
import random

with open("words.txt") as f:
    words = f.read().splitlines()

random_word = []
for key in random.choice(words):
    random_word.append(key)

word_len = len(random_word)

hidden_word = ["_"] * word_len

lives_left = 2

wrong_guess = []

def right_guess():
    if user_guess not in hidden_word:
        for index, char in enumerate(random_word):
            if char == user_guess:
                hidden_word[index] = char
        print(f" Guessed right: {' '.join(hidden_word)} next letter:")
    else:
        print("You already guessed this letter correct")


print(f'''
Welcome to the hangman
Today's word is: {" ".join(hidden_word)}
Try out with your first letter:
''')

print(random_word)

while random_word != hidden_word and lives_left > 0:
    user_guess = input()
    if user_guess in random_word:
        right_guess()
    elif user_guess not in random_word and lives_left > 0:
        if user_guess not in wrong_guess:
            lives_left -= 1
            wrong_guess.append(user_guess)
            if lives_left > 0:
                print(f"Wrong guess, try again! You have {lives_left} lives left")
            else:
                print("No lives left, you lost")
        else:
            print("You already guessed wrong this letter")

if random_word == hidden_word:
    print(f"You found the mistery word: '{''.join(hidden_word)}'")

```
