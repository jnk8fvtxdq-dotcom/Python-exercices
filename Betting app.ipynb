```python
bidders_dictionary = {}

def find_highest_bidder():
    highest_bidder = ""
    highest_bid = 0
    for bidder in bidders_dictionary:
        bids = bidders_dictionary[bidder]
        if bids > highest_bid:
            highest_bid = bids
            highest_bidder = bidder
    return f"The highest bidder is {highest_bidder} with ${highest_bid}"

while True:
    name = input("What's your name?")
    bid = int(input("What's your bid? $"))
    bidders_dictionary[name] = bid
    repeat = input("Continue? 'yes' or 'no'").lower()
    if repeat == "no":
        print(find_highest_bidder())
        break
```

```python
bidders_dictionary = {}

while True:
    name = input("What's your name?")
    bid = int(input("What's your bid? $"))
    bidders_dictionary[name] = bid
    repeat = input("Continue? 'yes' or 'no'").lower()
    if repeat == "no":
        break

highest_bidder = max(bidders_dictionary, key=bidders_dictionary.get)
highest_bid = bidders_dictionary[highest_bidder]

print(f"The highest bidder is {highest_bidder} with ${highest_bid}")
```
