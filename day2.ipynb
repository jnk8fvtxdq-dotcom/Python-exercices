```python
print("Welcome to the tip calculator!")
bill = float(input("What was the total bill? $"))
tip_pr = float(input("How much tip would you like to give? 10, 20 or 15?"))
pax = int(input("How many people to split the bill?"))
total = bill + bill * (tip_pr / 100)
total_per_pax = total / pax
print(f"Each person should pay: ${total_per_pax:.2f}")
```
