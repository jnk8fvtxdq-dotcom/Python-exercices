```python
weight = 85
height = 1.85

bmi = weight / (height ** 2)

if bmi < 18.5:
    print("Underweight")
elif bmi < 25:
    print("normal weight")
else:
    print("Overweight") 
```

Pizza delivery:
Size S: $15 M: $20 L: $25
Pepperoni: S: +$2 M and L: +$3 
Extra_cheese: +$1


```python
print("Welcome to Python Pizza Deliveries!")
size = input("What size of pizza do you want? S, M or L")
pepperoni = input("Do you want pepperoni on your pizza? Yes or No")
extra_cheese = input("Do you want extra cheese on your pizza? Yes or No")

size = size.upper()
pepperoni = pepperoni.upper()
extra_cheese = extra_cheese.upper()

bill = 0
if size == "S":
    bill += 15
elif size == "M":
    bill += 20
elif size == "L":
    bill += 25

if pepperoni == "YES" and size == "S":
    bill += 2
elif pepperoni == "YES":
    bill += 3
if extra_cheese == "YES":
        bill += 1


print(f"Your total bill is ${bill}")
```
